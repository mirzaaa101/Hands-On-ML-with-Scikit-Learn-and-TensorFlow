```python
from sklearn.datasets import make_regression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVR
import matplotlib.pyplot as plt
```

```python
# data
X, y = make_regression(n_samples=100, n_features=1, n_targets=1, random_state=42)
```

```python
# data distribution
plt.scatter(X, y, alpha=0.5)
plt.title("Linear Data")
plt.show()
```

```python
# model
svm_reg = make_pipeline(
    StandardScaler(),
    LinearSVR(epsilon=0.5, dual='auto', random_state=42)
)
```

```python
# training
svm_reg.fit(X,y)
```

```python
X_new = [[0.5],[-0.01]]
svm_reg.predict(X_new)
```

#### SVM Regression  With Kernel Trick To Tackle Nonlinear Data

```python
from sklearn.svm import SVR

#model
svm_poly_reg = make_pipeline(
    StandardScaler(),
    SVR(kernel="poly", degree=2, C=0.01, epsilon=0.1)
)

# training
svm_poly_reg.fit(X,y)
```
