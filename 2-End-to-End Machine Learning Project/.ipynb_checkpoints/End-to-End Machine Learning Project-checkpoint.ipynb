##### Assume that median income is more import to predict median_housing_prices. Since median_income holds continuous values, so, let's convert into categorical values. Let's categorize all the continues values into 5 categories

```python
df['income_category'] = pd.cut(
    df['median_income'], 
    bins = [0.0, 1.5, 3.0, 4.5, 6.0, np.inf],
    labels = [1,2,3,4,5]
)
df['income_category']
```

```python
# Let's visualize the income categories
df['income_category'].value_counts().sort_index().plot.bar(rot=0, grid=True)
plt.xlabel("Income Category")
plt.ylabel("Number of districts")
plt.show()
```

##### We're now ready to perform statified sampling while spliting data into train and test

```python
strat_train_set, strat_test_set = train_test_split(df, test_size=0.2, stratify=df['income_category'], random_state=42)

print(len(strat_train_set))
print(len(strat_test_set))
```

```python
# let's see if this works as expected
strat_test_set['income_category'].value_counts()/len(strat_test_set)
```

##### As you can see, the test set generated using stratified sampling has income category propotions almost identical to those in the full dataset, whereas the test set generate using purely random sampling is skewed. So, let's not use this column more and revert back into the previous version of the dataset.

```python
for _set in (strat_train_set, strat_test_set):
    _set.drop("income_category",axis=1, inplace=True)
```

#### Prepare the data for machine learning algorithms

```python
# split the attributs and target
X_train = strat_train_set.drop("median_house_value", axis=1)
y_train = strat_train_set["median_house_value"].copy()
```

##### Clean the data

```python
X_train["total_bedrooms"].isna().sum()
```

```python
# let's impute this attribute
median = df["total_bedrooms"].median()
X_train["total_bedrooms"].fillna(median, inplace=True)
```

##### Instead of using preceding code, we can also use Scikit-Learn class: 
    1. SimpleImputer
    2. KNNImputer: use meanof the K-nearest neighbors values for that feature
    3. InterativeImputer: use regression model to predict the missing values

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
```

##### Since median can only be computed againest numerical values, so let's create a copy of numerical attributes

```python
X_numerical = X_train.select_dtypes(include=[np.number])
X_numerical.dtypes
```

```python
# let's impute
imputer.fit(X_numerical)
```

```python
imputer.statistics_
```

```python
X_numerical.median().values
```

```python
X_imputed = imputer.transform(X_numerical)
X_imputed
```

##### Scikit-Learn transformers output Numpy arrays, so let's convert it into pandas DataFrame

```python
X_imputed = pd.DataFrame(X_imputed, columns=X_numerical.columns, index=X_numerical.index)
X_imputed.head()
```

#### Handling text and categorical attributes

```python
X_categorical = X_train.select_dtypes(include=['object'])
X_categorical.head(10)
```

#### Most ML algorithms prefer number intead of text or string. So we have to convert the categorical data into numurical, for thise we can also we Scikit-Learn classes like:
    1. OrdinalEncoder: prefers for orinal data(e.g., 'good'<'better'<'best')
    2. OneHotEncoder: prefers for categorical data like we have now
    3. pandas get_dummies() method: not as smarter as OneHotEncoder class

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
encoder.handle_unknown = "ignore"
encoded_X_categorical = encoder.fit_transform(X_categorical)
encoded_X_categorical
```

```python
encoded_X_categorical = encoded_X_categorical.toarray()
encoded_X_categorical
```

##### Again the output is in NumpY array, so let's convert into pandas DataFrame

```python
encoder.feature_names_in_
```

```python
encoder.get_feature_names_out()
```

```python
encoded_X_categorical_df = pd.DataFrame(
    encoded_X_categorical, columns=encoder.get_feature_names_out(), index=X_categorical.index
)
encoded_X_categorical_df
```

### Feature scaling and transformation
##### Most ML algorithms does not perform well when the input numerical values have different scales. So, we need to apply scaling trasformers:
    1. MinMaxScaler: is a normalization technique that is more sensitive to the outliers
    2. StandardScaler: is more smarter to handle outliers

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X_imputed)
X_scaled
```

```python
X_scaled = pd.DataFrame(X_scaled, columns=X_imputed.columns, index=X_imputed.index)
X_scaled
```

### Transformation Pipelines

Scikit-Learn provides a class called Pipeline that helps to do sequence of transformation task. The estimators must be transformers that they must have fit_transform() method. Let's do the imputation and scaling using Pipeline class.

```python
from sklearn.pipeline import Pipeline
from sklearn import set_config
from sklearn.linear_model import LinearRegression

steps = [(
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
)]
numerical_pipeline = Pipeline(steps)
sklearn.set_config(display="diagram")
numerical_pipeline
```

If you do not want to set name of the transformers you can use make_pipeline() method, is takes positional arguments, and creates Pipeline using the name of the transformers class name in lowercase with no underscore (e.g, 'simpleimputer').

```python
from sklearn.pipeline import make_pipeline

numerical_pipeline = make_pipeline(
    SimpleImputer(strategy="median"), StandardScaler()
)

X_numerical_predpared = numerical_pipeline.fit_transform(X_scaled)
X_numerical_predpared
```

```python
# We need to convert the output into pandas DataFrame
X_numerical_predpared = pd.DataFrame(
    X_numerical_predpared, columns=numerical_pipeline.get_feature_names_out, index=X_scaled.index
)
X_numerical_predpared
```
