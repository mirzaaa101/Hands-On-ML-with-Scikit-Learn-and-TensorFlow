```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
```

```python
np.random.seed(42)
X_quad = np.random.rand(200, 1) - 0.5
y_quad = X_quad ** 2 + 0.025 * np.random.rand(200,1)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_quad, y_quad, alpha=0.5)
plt.show()
```

```python
tree_reg = DecisionTreeRegressor(max_depth=2, min_samples_leaf=10, random_state=42)
tree_reg.fit(X_quad, y_quad)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(
    tree_reg,
    out_file="reg_tree.dot",
    feature_names = ['X'],
    class_names=['y'],
    rounded=True,
    filled=True
)
```

```python
from graphviz import Source
Source.from_file('reg_tree.dot')
```
