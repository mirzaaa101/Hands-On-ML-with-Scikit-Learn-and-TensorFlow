The Keras preprocessing API includes three image preprocessing layers:

1. `tf.keras.layers.Resizing`: resizes the input images to the desired size.
2. `tf.keras.layers.Rescaling`: rescales the pixel values.
3. `tf.keras.layers.CenterCrop`: keeps only the center patch of the desired height and width.

```python
import tensorflow as tf
from sklearn.datasets import load_sample_images

images = load_sample_images()["images"]
```

```python
import matplotlib.pyplot as plt

def show_image(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
show_image(images[0])
```

```python
show_image(images[1])
```

```python
images[0].shape
```

```python
crop_image_layer = tf.keras.layers.CenterCrop(height=300, width=300)
cropped_images = crop_image_layer(images)
```

```python
show_image(cropped_images[1])
```

Keras also includes several layers for `Data Augmentation` such as RandomCrop, RandomFlip, RandomTranslation, RandomRotation, RandomHeight, RandomWidth, RandomContrast.
